```python
#Import the necessary libraries to enable us perform our analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Define the data path
df = pd.read_csv('C:/Diabatic_data/extra3/advection/temp_tendency.csv')
df.columns     ## Check the column headers
```

```python
fig = plt.figure(figsize=(10,6))
df1 = pd.DataFrame(df,columns=['Years','SA_08_3'])
df2 = pd.DataFrame(df,columns=['Years','SAF_DHR'])
df3 = pd.DataFrame(df,columns=['Years','SA_hroz'])
df4 = pd.DataFrame(df,columns=['Years','SA_vert'])
df5 = pd.DataFrame(df,columns=['Years','SA_adia'])
plt.plot(df1['Years'], df1['SA_08_3'], color='black', marker='x')
plt.plot(df2['Years'], df2['SAF_DHR'], color='green', marker='x')
plt.plot(df3['Years'], df3['SA_hroz'], color='orange', marker='x')
plt.plot(df4['Years'], df4['SA_vert'], color='red', marker='x')
plt.plot(df5['Years'], df5['SA_adia'], color='purple', marker='x')
plt.yticks(np.arange(-3, 4, 1), fontsize=12, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.axhline(y = 0, color = 'black', linestyle = '--', marker='>')
plt.axvline(x = 1981, color = 'blue', linestyle = '--', marker='>')
plt.axvline(x = 2000, color = 'blue', linestyle = '--', marker='>')
#plt.axhline(x = 0, color = 'black', linestyle = '--', marker='>')
plt.title('SAF annual anomaly of local temperature tendency terms and PHEs (1961-2021)', fontsize=12, fontweight='bold')
plt.xlabel('Years', fontsize=12,fontweight='bold' )
plt.ylabel('temp tendency terms', fontsize=14, fontweight='bold')
plt.legend(['Southern African PHEs','Diabating heating','Horizontal thermal advection','Vertical thermal advection','Adiabatic expansion/compression',], loc = 'upper left',fontsize=9)
plt.text(1965, -2.5, 'Inactive PHEs',fontsize = 12, color ='blue')
plt.text(1985, -2.5, 'Active PHEs',fontsize = 12, color ='blue')
plt.text(2005, -2.5, 'Inactive PHEs',fontsize = 12, color ='blue')
plt.savefig('C:/Diabatic_data/extra3/advection/temp_tendency term.png')
plt.show()
```

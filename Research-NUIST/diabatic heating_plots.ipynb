### 4 Panel plot for diabatic heating rate fields and PHE by Ekwacu Samuel

```python
#Import the necessary libraries to enable us perform our analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Define the data path
df = pd.read_csv('C:/Diabatic_data/MONTHS/dfhr/4panel_dfhr.csv')
df.columns     ## Check the column headers
```

```python
scipy.stats.pearsonr(df['NA_08_3'], df['NAF_sum'])    # Pearson's r
#scipy.stats.spearmanr(df['NA_08_3'], df['EA_08_3'])   # Spearman's rho
#scipy.stats.kendalltau(df['NA_08_3'], df['EA_08_3'])  # Kendall's tau
```

```python
fig, ax = plt.subplots(2,2, figsize =(16,10))

#plot 1
ax[0,0].plot(df['Years'], df['NA_08_3'],color='black', marker='x')
ax[0,0].plot(df['Years'], df['NAF_sum'],color='green', marker='x')
ax[0,0].plot(df['Years'], df['NAF_lwhr'],color='orange', marker='x')
ax[0,0].plot(df['Years'], df['NAF_swhr'],color='red', marker='x')
ax[0,0].set_title('NAF Annual anomaly of DFHRs and PHEs (1961-2021)', fontsize=10, fontweight='bold')
ax[0,0].set_ylabel('Anomaly', fontsize=10, fontweight='bold')
#ax[0,0].set_yticks(np.arange(0, 22, 2), fontsize=10, fontweight='bold')
ax[0,0].axhline(y = 0, color = 'black', linestyle = '--', marker='>')
ax[0,0].legend(['PHE','DFHR_sum','LWHR','SWHR'], loc = 'upper left',fontsize=9)
#ax[0,0].text(2019, 19, '(a)',fontsize = 14, color ='brown', fontweight='bold')
#ax[0,0].grid(True)

#plot 2
ax[0,1].plot(df['Years'], df['EA_08_3'],color='black', marker='x')
ax[0,1].plot(df['Years'], df['EAF_sum'],color='green', marker='x')
ax[0,1].plot(df['Years'], df['EAF_lwhr'],color='orange', marker='x')
ax[0,1].plot(df['Years'], df['EAF_swhr'],color='red', marker='x')
ax[0,1].set_title('EAF Annual anomaly of DFHRs and PHEs (1961-2021)', fontsize=10, fontweight='bold')
ax[0,1].set_ylabel('Anomaly', fontsize=10, fontweight='bold')
#ax[0,1].set_yticks(np.arange(0, 22, 2), fontsize=10, fontweight='bold')
ax[0,1].axhline(y = 0, color = 'black', linestyle = '--', marker='>')
ax[0,1].legend(['PHE','DFHR_sum','LWHR','SWHR'], loc = 'upper left',fontsize=9)
#ax[0,1].text(2010, 18, '(b)',fontsize = 14, color ='brown',fontweight='bold')
#ax[0,1].grid(True)

#plot 3
ax[1,0].plot(df['Years'], df['WA_08_3'],color='black', marker='x')
ax[1,0].plot(df['Years'], df['WAF_sum'],color='green', marker='x')
ax[1,0].plot(df['Years'], df['WAF_lwhr'],color='orange', marker='x')
ax[1,0].plot(df['Years'], df['WAF_swhr'],color='red', marker='x')
ax[1,0].set_title('WAF Annual anomaly of DFHRs and PHEs (1961-2021)', fontsize=10, fontweight='bold')
ax[1,0].set_ylabel('Anomaly', fontsize=10, fontweight='bold')
ax[1,0].set_xlabel('Years', fontsize=10, fontweight='bold')
plt.xlabel('Years', fontsize=12, fontweight='bold')
#ax[1,0].set_yticks(np.arange(0, 22, 2), fontsize=10,fontweight='bold')
ax[1,0].axhline(y = 0, color = 'black', linestyle = '--', marker='>')
ax[1,0].legend(['PHE','DFHR_sum','LWHR','SWHR'], loc = 'upper left',fontsize=9)
#ax[1,0].text(2005, 19, '(c)',fontsize = 14, color ='brown',fontweight='bold')
#ax[1,0].grid(True)

#plot 4
ax[1,1].plot(df['Years'], df['SA_08_3'],color='black', marker='x' )
ax[1,1].plot(df['Years'], df['SAF_sum'],color='green', marker='x' )
ax[1,1].plot(df['Years'], df['SAF_lwhr'],color='orange', marker='x')
ax[1,1].plot(df['Years'], df['SAF_swhr'],color='red', marker='x' )
ax[1,1].set_title('SAF Annual anomaly of DFHRs and PHEs (1961-2021)', fontsize=10, fontweight='bold')
ax[1,1].set_ylabel('Anomaly', fontsize=10, fontweight='bold')
ax[1,1].set_xlabel('Years', fontsize=12, fontweight='bold')
#ax[1,1].set_yticks(np.arange(0, 20, 2),fontsize=10, fontweight='bold')
ax[1,1].axhline(y = 0, color = 'black', linestyle = '--', marker='>')
ax[1,1].legend(['PHE','DFHR_sum','LWHR','SWHR',], loc = 'upper left',fontsize=9)
#ax[1,1].text(2019, 18, '(d)',fontsize = 14, color ='brown',fontweight='bold')
#ax[1,1].set_xticks(fontsize=8, fontweight='bold')
#ax[1,1].grid(True)
plt.savefig('C:/Diabatic_data/MONTHS/dfhr/DFHRs.png', dpi=300)
plt.show()
```

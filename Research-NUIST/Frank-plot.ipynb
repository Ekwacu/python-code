### To Interpret the output of the Test:
1. trend: This tells the trend. Possible output includes increasing, decreasing, or no trend.
2. h: True if trend is present. False if no trend is present.
3. p: The p-value of the test.
4. z: The normalize test statistic.
5. Tau: Kendall Tau.
6. s: Mann-Kendals score
7. var_s: Variance S
8. slope: Theil-Sen estimator/slope
9. intercept: Intercept of Kendall-Theil Robust Line
10. The main value were interested in is the p-value, which tells us whether or not there is a statistically significant trend in the data.
11. For p-value <0.05 implies a significant trend exist while for P-values >0.05 implies no significant trend exist in the time series data.

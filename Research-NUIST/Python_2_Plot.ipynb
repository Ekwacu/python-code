### python script to plot cyclone events by Dinis

```python
## Import the necessary libraries to enable us perform our analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib.inline
```

```python
#home_dir = 'D:/Dinis_Timeseries/'
df = pd.read_csv('C:/JRA55_temp/Dinis/Final_Timeseries/Landfall.csv')
df.columns
#df.head(20)
```

```python
df.describe() 
```

```python
# calculating exponential moving average
# using .ewm(span).mean() , with window size = 10
df['SA_LF'] =df['Number'].ewm(span=10).mean()
# printing Dataframe
#df['SA_LF']
```

```python
#MK test
import pymannkendall as mk
d0 = mk.original_test(df['Number'])
print(d0)
```

```python
fig = plt.figure(figsize=(8,5))
df2 = pd.DataFrame(df,columns=['Year','SA_LF'])
plt.plot(df2['Year'], df2['SA_LF'], color='red')
df1 = pd.DataFrame(df,columns=['Year','Number'])
plt.bar(df1['Year'], df1['Number'], color='blue')#, marker='o')
plt.yticks(np.arange(0, 4, 0.5), fontsize=12, fontweight='normal')
plt.xticks(fontsize=12, fontweight='normal')
plt.xticks(np.arange(1980, 2022, 2)) 
plt.xticks(rotation=75)
plt.title('Annual number of Landfall Cyclones over Mozambique', fontsize=12, fontweight='bold')
plt.xlabel('Years', fontsize=12, fontweight='bold')
plt.ylabel('Number of Cyclones', fontsize=12, fontweight='bold')
plt.axhline(y = 0.46, color = 'green', linestyle = '--', marker='>')
z = np.polyfit(df['Year'], df['Number'], 1)
p = np.poly1d(z)
plt.plot(df['Year'],p(df['Year']),"black")
plt.legend(['1O year runnung mean','Mean','Trend', 'Cyclone number'], loc = 'upper left')
plt.text(1996, 2.7, 'MK trend Pvalue =0.7061',fontsize = 12, color ='purple')
#plt.grid(True)
plt.savefig('C:/JRA55_temp/Dinis/Final_Timeseries/Landfalf.png')
plt.show()
```

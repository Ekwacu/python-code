### Calculating Exponential Moving Average by Ekwacu Samuel

```python
#Import the necessary libraries to enable us perform our analysis
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
home_dir = 'D:/Folder_2023/Research/Africa_Research/Hotspell_Africa/extracted_data/cluster_regions/'
df = pd.read_csv(home_dir +'Africa1_region-PHE.csv')
df.columns     ## Check the column headers
df.head(5)
```

```python
# calculating exponential moving average
# using .ewm(span).mean() , with window size = 10
df['NA_EWMA10'] =df['NA_08_3'].ewm(span=10).mean()
df['EA_EWMA10'] =df['EA_08_3'].ewm(span=10).mean()
df['WA_EWMA10'] =df['WA_08_3'].ewm(span=10).mean()
df['SA_EWMA10'] =df['SA_08_3'].ewm(span=10).mean()
df['Af_EWMA10'] =df['Af_08_3'].ewm(span=10).mean()
#For PHEs Days
df['NA_EWMA10D'] =df['NA_phe'].ewm(span=10).mean()
df['EA_EWMA10D'] =df['EA_phe'].ewm(span=10).mean()
df['WA_EWMA10D'] =df['WA_phe'].ewm(span=10).mean()
df['SA_EWMA10D'] =df['SA_phe'].ewm(span=10).mean()
df['Af_EWMA10D'] =df['Af_phe'].ewm(span=10).mean()
# printing Dataframe
#df['NA_EWMA10']
#df['EA_EWMA10']
#df['WA_EWMA10']
#df['SA_EWMA10']
#df['Af_EWMA10']
```

```python
fig = plt.figure(figsize=(8,4))
# plotting phe and exponential
# moving averages of 10 years
# using .plot() method
df12 = pd.DataFrame(df,columns=['Years','NA_EWMA10'])
df13 = pd.DataFrame(df,columns=['Years','NA_08_3'])
plt.plot(df12['Years'], df12['NA_EWMA10'], color='orange')#, marker='o')
plt.plot(df13['Years'], df13['NA_08_3'], color='green', marker='o')
plt.axhline(y = 10.72131148, color = 'blue', linestyle = '--', marker='>')
z = np.polyfit(df['Years'], df['NA_08_3'], 1)
p = np.poly1d(z)
plt.plot(df['Years'],p(df['Years']),"r--")
plt.title('Annual mean PHEs-criteria 0.8 degC (1961-2021)', fontsize=14, fontweight='bold')
plt.yticks(np.arange(0, 22, 2), fontsize=10, fontweight='bold')
plt.xticks(fontsize=10, fontweight='bold')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Hot Spell events', fontsize=14)
#plt.grid(True)
plt.legend(['10 Year moving average', 'Northern Africa yearly PHE','61-Year PHE mean','Trend (1.5 PHEs/decade)'])
plt.savefig(home_dir + 'Hot Spell-single-plot.png')
```

### To Interpret the output of the Test:

1. trend: This tells the trend. Possible output includes increasing, decreasing, or no trend.
2. h: True if trend is present. False if no trend is present.
3. p: The p-value of the test.
4. z: The normalize test statistic.
5. Tau: Kendall Tau.
6. s: Mann-Kendal’s score
7. var_s: Variance S
8. slope: Theil-Sen estimator/slope
9. intercept: Intercept of Kendall-Theil Robust Line
10. The main value we’re interested in is the p-value, which tells us whether or not there is a statistically significant trend in the data.
11. For p-value <0.05 implies a significant trend exist while for P-values >0.05 implies no significant trend exist in the time series data.
 

#### NOTES ON MANN- KENDALL TEST
#### What is the Mann-Kendall Test ?
1. The Mann-Kendall Trend Test (sometimes called the MK test) is used to analyze time series data for consistently increasing or decreasing trends (monotonic trends). It is a non-parametric test, which means it works for all distributions (i.e. data doesn't have to meet the assumption of normality), but data should have no serial correlation. If the data has a serial correlation, it could affect in significant level (p-value). It could lead to misinterpretation. To overcome this problem, researchers proposed several modified Mann-Kendall tests (Hamed and Rao Modified MK Test, Yue and Wang Modified MK Test, Modified MK test using Pre-Whitening method, etc.). Seasonal Mann-Kendall test also developed to remove the effect of seasonality.

2. Mann-Kendall Test is a powerful trend test, so several others modified Mann-Kendall tests like Multivariate MK Test, Regional MK Test, Correlated MK test, Partial MK Test, etc. were developed for the spacial condition. pyMannkendal is a pure Python implementation of non-parametric Mann-Kendall trend analysis, which bring together almost all types of Mann-Kendall Test. Currently, this package has 11 Mann-Kendall Tests and 2 sen's slope estimator function. Brief description of functions are below:

3. Original Mann-Kendall test (original_test): Original Mann-Kendall test is a nonparametric test, which does not consider serial correlation or seasonal effects.

4. Hamed and Rao Modified MK Test (hamed_rao_modification_test): This modified MK test proposed by Hamed and Rao (1998) to address serial autocorrelation issues. They suggested a variance correction approach to improve trend analysis. User can consider first n significant lag by insert lag number in this function. By default, it considered all significant lags.

5. Yue and Wang Modified MK Test (yue_wang_modification_test): This is also a variance correction method for considered serial autocorrelation proposed by Yue, S., & Wang, C. Y. (2004). User can also set their desired significant n lags for the calculation.

6. Modified MK test using Pre-Whitening method (pre_whitening_modification_test): This test suggested by Yue and Wang (2002) to using Pre-Whitening the time series before the application of trend test.

7. Modified MK test using Trend free Pre-Whitening method (trend_free_pre_whitening_modification_test): This test also proposed by Yue and Wang (2002) to remove trend component and then Pre-Whitening the time series before application of trend test.

8. Multivariate MK Test (multivariate_test): This is an MK test for multiple parameters proposed by Hirsch (1982). He used this method for seasonal mk test, where he considered every month as a parameter.

9. Seasonal MK Test (seasonal_test): For seasonal time series data, Hirsch, R.M., Slack, J.R. and Smith, R.A. (1982) proposed this test to calculate the seasonal trend.

10. Regional MK Test (regional_test): Based onHirsch (1982) proposed seasonal mk test, Helsel, D.R. and Frans, L.M., (2006) suggest regional mk test to calculate the overall trend in a regional scale.

11. Correlated Multivariate MK Test (correlated_multivariate_test): This multivariate mk test proposed by Hipel (1994) where the parameters are correlated.

12. Correlated Seasonal MK Test (correlated_seasonal_test): This method proposed by Hipel (1994) used, when time series significantly correlated with the preceding one or more months/seasons.

13. Partial MK Test (partial_test): In a real event, many factors are affecting the main studied response parameter, which can bias the trend results. To overcome this problem, Libiseller (2002) proposed this partial mk test. It required two parameters as input, where, one is response parameter and other is an independent parameter.

14. Theil-Sen's Slope Estimator (sens_slope): This method proposed by Theil (1950) and Sen (1968) to estimate the magnitude of the monotonic trend. Intercept is calculate using Conover, W.J. (1980) method.

15. Seasonal Theil-Sen's Slope Estimator (seasonal_sens_slope): This method proposed by Hipel (1994) to estimate the magnitude of the monotonic trend, when data has seasonal effects. Intercept is calculate using Conover, W.J. (1980) method.

#### Function details: All Mann-Kendall test functions have almost similar input parameters. Those are:

1. x: a vector (list, numpy array or pandas series) data
2. alpha: significance level (0.05 is the default)
3. lag: No. of First Significant Lags (Only available in hamed_rao_modification_test and yue_wang_modification_test)
4. period: seasonal cycle. For monthly data it is 12, weekly data it is 52 (Only available in seasonal tests)

#### And all Mann-Kendall tests return a named tuple which contained:

1. trend: tells the trend (increasing, decreasing or no trend)
2. h: True (if trend is present) or False (if the trend is absence)
3. p: p-value of the significance test
4. z: normalized test statistics
5. Tau: Kendall Tau
6. s: Mann-Kendal's score
7. var_s: Variance S
8. slope: Theil-Sen estimator/slope
9. intercept: intercept of Kendall-Theil Robust Line, for seasonal test, full period cycle consider as unit time step
10. sen's slope function required data vector. seasonal sen's slope also has optional input period, which by the default value is 12. Both sen's slope function return only slope value.

```python
## Import the necessary libraries to enable us perform our analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Define the data path

home_dir = 'G:/Folder_2022/Research/'
```

```python
## Read in the data using the appropriate Library and by defining the correct path

hotspell = pd.read_csv(home_dir +'Hotspell_events-Tmax.csv')
#hotspell             ## Prints out a sample of the data
#hotspell.info()      ## describe your data
hotspell.columns     ## Check the column headers
#hotspell.shape       ## Check the number of rows and columns
#hotspell.describe()  ## Get summary statistics about your data
#hotspell[["Hotspells12_BR"]].describe()  ## Summary statistics on particular columns as a dataframe
#hotspell["Hotspells12_BR"].describe()  ## Summary statistics on particular columns as a series
#hotspell.sort_values(by="Hotspells12_BR", ascending = False) ## Sort the data in descending order
```

```python
col_mean = pd.DataFrame(hotspell.mean(axis=0))
mon_nms = col_mean.index[1:20];mon_nms
#col_mean[0][1:13]
yr_col = hotspell.columns[0];yr_col
yrs = hotspell[yr_col]; yrs #displays the year in the data
Br_hotspell = hotspell.columns[1];Br_hotspell
Br_12 = hotspell[Br_hotspell]; Br_12
Br_hotspell = hotspell.columns[2];Br_hotspell
Br_15 = hotspell[Br_hotspell]
#Br_15
```

###### Plot a Line chart in Python using Matplotlib

```python
plt.plot(yrs, Br_12, 'o-g')
plt.plot(yrs, Br_15, 'o-b')
plt.xlabel("Year", fontsize=14)
plt.ylabel("hotspell events", fontsize=14)
plt.title("Annual Number of hotspell events for Burundi", fontsize=14)
#legend 
plt.legend(['criteria 1.2', 'Criteria 1.5'])
#plt.grid(True)
plt.savefig(home_dir + 'python_plots/' + 'hotspell_Burundi-tmax.png')
plt.show()
```

###### Create a Line Chart in Python with Pandas DataFrame

```python
df = pd.DataFrame(hotspell,columns=['Year','Hotspells12_BR'])
df1 = pd.DataFrame(hotspell,columns=['Year','Hotspells15_BR'])
plt.plot(df['Year'], df['Hotspells12_BR'], color='red', marker='o')
plt.plot(df['Year'], df1['Hotspells15_BR'], color='green', marker='o')
plt.title('Annual Number of hotspell events for Burundi', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('hotspell events', fontsize=14)
#legend 
plt.legend(['criteria 1.2', 'Criteria 1.5'])
#plt.grid(True)
plt.savefig(home_dir + 'python_plots/' + 'hotspell_Burundi1-tmax.png')
plt.show()
```

```python
#Make an average of the wind variables

#For SON
wspd_son = np.mean(np.sqrt(np.add(np.square(ond.u10),np.square(ond.v10))),axis=0)
u_son = np.mean(ond.u10,axis=0)
v_son = np.mean(ond.v10,axis=0)
#u_son
```

```python
#setting up the quiver arguments (these display the wind vectors on the plot)
xx_wind = wspd_son.longitude.values
yy_wind = wspd_son.latitude.values
X_wind,Y_wind =np.meshgrid(xx_wind, yy_wind)
U_wind = u_son.data
V_wind = v_son.data
```
